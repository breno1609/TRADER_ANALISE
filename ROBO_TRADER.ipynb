```python
#CONCATENANDO AS PLANILHAS
df = [df1, df2, df3]
df = pd.concat(df) #VAI CONCATENAR TODOS OS ELEMNTOS QUE ESTÃO NA LISTA DF
```

```python
df
```

```python
#SELECIONANDO AS COLUNAS QUE VÃO SER UTILIZADAS NA BASE DE DADOS
df = df[['Div', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 
         'FTR', 'HTHG', 'HTAG', 'HTR', 'AS', 
         'HST', 'AST', 'HC', 'AC', 'HF', 'AF', 
         'B365H', 'B365D', 'B365A']]
```

```python
#RENOMEAR ASA COLUNAS RESULTADOD JOGO(H = VITORIA CASA, D = EMPATE, A = VITORIA FORA)
df.columns = ['Divisao', 'Mandante', 'Visitante', 'GolsMandante', 'GolsVisitantes', 'ResultadoJogo', 
              'GolsCasaIntervalo', 'GolsForaIntervalo', 'ResultadoIntervalo', 'ChuteForaCasa', 
              'ChuteAlvoCasa', 'ChuteAlvoFora', 'CantoCasa', 'CantoFora', 'FaltasCasa', 'FaltasFora', 
              'ProbVitCasa', 'ProbEmpate', 'ProbVitFora']

df.reset_index(inplace=True, drop=True)
df.index = df.index.set_names(['Nº'])
df = df.rename(index=lambda x: x + 1)
df = df.dropna()
```

```python
#DADOS DE PONTOS
def Pontos_H(ResultadoJogo): #TIME MANDANTE
    if ResultadoJogo == 'H':
        Pontos = 3
    elif ResultadoJogo == 'D':
        Pontos = 1
    else:
        Pontos = 0
        
def Pontos_A(ResultadoJogo): #TIME VISITANTE
    if ResultadoJogo == 'A':
        Pontos = 3
    elif ResultadoJogo == 'D':
        Pontos = 1
    else:
        Pontos = 0
    return Pontos

df['Pontos_casa'] = df.apply(lambda row: Pontos_H(row['ResultadoJogo']), axis=1)
df['Pontos_fora'] = df.apply(lambda row: Pontos_A(row['ResultadoJogo']), axis=1)
```

```python
#DADOS DE GOL

df['TotalGolsPrimTempo'] = df.apply(lambda row: (row['GolsCasaIntervalo'] + row['GolsForaIntervalo']), axis=1) 

df['TotalGolsJogo'] = df.apply(lambda row: (row['GolsMandante'] + row['GolsVisitantes']), axis=1) 

df['MaisDe0Gol1Tmp'] = df.apply(lambda row: 1 if (row['TotalGolsPrimTempo'] > 0) else 0, axis=1) #1 VERDADEIRO E 0 FALSO + DE 0 GOLS NO PT

df['MaisDe0GolJogo'] = df.apply(lambda row: 1 if (row['TotalGolsJogo'] > 0) else 0, axis=1) # + 0 GOLS JOGO

df['MaisDe1GolJogo'] = df.apply(lambda row: 1 if (row['TotalGolsJogo'] > 1) else 0, axis=1)  # + 1 GOL NO JOGO

df['Mais2GolJogo'] = df.apply(lambda row: 1 if (row['TotalGolsJogo'] > 2) else 0, axis=1) #+ 2 GOLS NO JOGO

df['BTTS'] = df.apply(lambda row: 1 if (row['GolsMandante'] > 0 and row['GolsVisitantes'] > 0) else 0, axis=1) #AMBOS MARCAM
```

```python
#PORCENTAGENS
df['Porc_Over05HT_Casa'] = (df.groupby('Mandante')['MaisDe0Gol1Tmp'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
df['Porc_Over05HT_Visitante'] = (df.groupby('Visitante')['MaisDe0Gol1Tmp'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)

df['Porc_Over05HT_Casa'] = (df.groupby('Mandante')['MaisDe0GolJogo'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
df['Porc_Over05HT_Visitante'] = (df.groupby('Visitante')['MaisDe0GolJogo'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)

df['Porc_Over05HT_Casa'] = (df.groupby('Mandante')['MaisDe0GolJogo'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
df['Porc_Over05HT_Visitante'] = (df.groupby('Visitante')['MaisDe0GolJogo'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)

df['Porc_Over05HT_Casa'] = (df.groupby('Mandante')['Mais2GolJogo'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
df['Porc_Over05HT_Visitante'] = (df.groupby('Visitante')['Mais2GolJogo'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)

df['Porc_BTTS_Casa'] = (df.groupby('Mandante')['BTTS'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
df['Porc_BTTS_Visitante'] = (df.groupby('Visitante')['BTTS'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100)
```

```python
#ESTATISTICA PONTOS
# Estatísticas de Pontos

# Média
df['Media_Pontos_casa'] = df.groupby('Mandante')['Pontos_casa'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)
df['Media_Pontos_fora'] = df.groupby('Visitante')['Pontos_fora'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)

# Desvio Padrão
df['DP_Pontos_casa'] = df.groupby('Mandante')['Pontos_casa'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)
df['DP_Pontos_fora'] = df.groupby('Visitante')['Pontos_fora'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)

# Coeficiente de Variação
df['CV_Pontos_casa'] = df['DP_Pontos_casa'] / df['Media_Pontos_casa']
df['CV_Pontos_fora'] = df['DP_Pontos_fora'] / df['Media_Pontos_fora']
```

```python
# Estatísticas de Gols

# Média
df['Media_Gols_Feitos_Home'] = df.groupby('Mandante')['GolsMandante'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)
df['Media_Gols_Sofridos_Home'] = df.groupby('Mandante')['GolsVisitantes'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)

df['Media_Gols_Feitos_Away'] = df.groupby('Visitante')['GolsVisitantes'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)
df['Media_Gols_Sofridos_Away'] = df.groupby('Visitante')['GolsMandante'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True)

# Desvio Padrão
df['DP_Gols_Feitos_Home'] = df.groupby('Mandante')['GolsMandante'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)
df['DP_Gols_Sofridos_Home'] = df.groupby('Mandante')['GolsVisitantes'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)

df['DP_Gols_Feitos_Away'] = df.groupby('Visitante')['GolsVisitantes'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)
df['DP_Gols_Sofridos_Away'] = df.groupby('Visitante')['GolsMandante'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True)

# Coeficiente de Variação
df['CV_Gols_Feitos_Home'] = (df['DP_Gols_Feitos_Home'] / df['Media_Gols_Feitos_Home'])
df['CV_Gols_Sofridos_Home'] = (df['DP_Gols_Sofridos_Home'] / df['Media_Gols_Sofridos_Home'])

df['CV_Gols_Feitos_Away'] = (df['DP_Gols_Feitos_Away'] / df['Media_Gols_Feitos_Away'])
df['CV_Gols_Sofridos_Away'] = (df['DP_Gols_Sofridos_Away'] / df['Media_Gols_Sofridos_Away'])
```

```python
# Probabilidades
df['p(H)'] = df.apply(lambda row: (1/ row['ProbVitCasa']), axis=1)
df['p(D)'] = df.apply(lambda row: (1/ row['ProbEmpate']), axis=1)
df['p(A)'] = df.apply(lambda row: (1/ row['ProbVitFora']), axis=1)
```
